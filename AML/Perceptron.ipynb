# Perceptron algorithm - Widrow-Hoff law 

OR, AND function


1. my code - repeat for a given numbers of iterations


```python
# training base S(X,c)
x=[[0,0], [0,1], [1,0], [1,1]] # input
c=[0,1,1,1] # result of OR 
# c=[0,0,1,1] # result of AND
o=[-1,-1,-1,-1] # result of training

# set randomly initial weights (W, b)
w=[-1, 1] # weight
b=0.5 # bias
n=0.2 # learning rate (클수록 빨리 예측)

# training
# repeat for a given number of iterations
epoch = 10
for a in range (epoch):
  print("< " + str(a) + " epoch >\n")

  # take each example (X,c) from S
  for i in range(len(x)):
    print("(x0, x1, c) = (" + str(x[i][0]) + ", " + str(x[i][1]) + ", " + str(c[i]) + ")")

    # calculate the output o of the perceptron based on the input x
    v=w[0]*x[i][0] + w[1]*x[i][1] + b
    if v > 0:
      o[i]=1
    else:
      o[i]=0

    print("v = " + str(v) + ", o = " + str(o))

    # update the weights (W, b)
    w[0] = w[0] + n*(c[i]-o[i]) * x[i][0]
    w[1] = w[1] + n*(c[i]-o[i]) * x[i][1]
    b = b+n*(c[i]-o[i])

    print("(w0, w1, b) = (" + str(w[0]) + ", " + str(w[1]) + ", " + str(b) + ")\n")

# test
print("< test >\n")

t=[[0,0], [0,1], [1,0], [1,1]] # test value

for i in range(4):
    if t[i][0] == x[i][0]:
      if t[i][1] == x[i][1]:
        print("input = [" + str(t[i][0]) + ", " + str(t[i][1]) + "], result = " + str(o[i]))
```


2. Solution - repeat until all examples are well predicted


```python
import numpy as np

def predicton(x,w,c):
    loss = 0
    for i in range(x.shape[0]):
        v = x[i].dot(w)+b
        # dot function 
        # a[x,y], b[z,w]
        # a.dot(b) = xz + yw

        if(v >=0):
            o=1
        else:
            o=0
        loss = loss + abs(c[i]-o)
        # abs: return the absolute value of the argument(절대값)
        # abs(c[i]-o): c와 o의 차이

    print('Obtained loss: {}'.format(loss))
        
    if loss == 0:
        return True
    else :
        return False
    
        
x = np.array(([0,0],[0,1],[1,0],[1,1])) # inputs
c = np.array((0,1,1,1)) # result of OR
w = np.array([-1, 1]) # weights
b = 0.5 # bias
lr = 0.2 # Learning rate: 0.002, 0.02 0.2 2
epochs = 1


while True: # true가 될때까지 (loss = 0이 될때까지)
    print('epoch: {}'.format(epochs))
    
    for i in range(x.shape[0]): # for a data point entry
        v = x[i].dot(w)+b # calculate the transfer function
        if(v > 0): # apply the activation function 'heavyside' to get output o
            o=1
        else:
            o=0
        # update weights and bias
        w = w + lr*(c[i]-o)*x[i]
        b = b + lr*(c[i]-o)
    
    if (predicton(x,w,c)): # we finished one epoch, we check if the neuron predicts outputs well
        break
    epochs = epochs+1


print ('\nSolved in {} epochs!'.format(epochs))
print ('final weight w {}'.format(w))
print ('final bias b {}'.format(b))
```
