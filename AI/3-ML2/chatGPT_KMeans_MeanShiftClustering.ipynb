# K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# load data from "kmeans.csv"
data = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/JUNIA/AI/kmeans.csv')

# initialize k-means model
kmeans = KMeans(n_clusters=3)

# fit the model to the data
kmeans.fit(data)

# predict the cluster labels for each point
labels = kmeans.predict(data)

# plot the data points with their cluster labels
plt.scatter(data[:,0], data[:,1], c=labels)
plt.show()
```

# Mean Shift

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift

# load data from "kmeans.csv"
data = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/JUNIA/AI/kmeans.csv')

# initialize mean shift model
ms = MeanShift()

# fit the model to the data
ms.fit(data)

# predict the cluster labels for each point
labels = ms.predict(data)

# plot the data points with their cluster labels
plt.scatter(data[:,0], data[:,1], c=labels)
plt.show()
```
